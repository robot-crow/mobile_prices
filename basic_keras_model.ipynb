```python
'''A simple keras model for mobile prices based upon charactersistics of phone. Data taken from Kaggle.'''
```

```python
import numpy as np
import pandas as pd

import os

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense

from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
cwd = os.getcwd()
data_csv = os.path.join(cwd, 'data/data.csv')
```

```python
df = pd.read_csv(data_csv)
```

```python
pd.set_option('display.max_columns',15)
```

```python
df
```

```python
def get_data(df: pd.DataFrame) -> tuple[np.ndarray, np.ndarray]:
    # as numpy arrays
    # X = everything minus target
    # y = target

    
    X = df.iloc[:,:20].values
    y = df.iloc[:,20:21].values

    # scale & encode
    sc = StandardScaler()
    ohe = OneHotEncoder()
    
    X = sc.fit_transform(X)
    y = ohe.fit_transform(y).toarray()

    # one time arbitrary split. We just want to get going here, not productionise
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
    
    return X_train, X_test, y_train, y_test
    
```

```python
def get_model() -> keras.models.Sequential:
    model = Sequential()
    model.add(Dense(16, input_dim=20, activation='relu'))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(4, activation='softmax'))

    model.compile(loss='categorical_crossentropy', 
                  optimizer='adam', 
                  metrics=['accuracy'])

    return model
```

```python
X_train, X_test, y_train, y_test = get_data(df)
```

```python
# len(X_train[0])
# len(X_train)
# len(X_test[0])
# len(X_test)
```

```python
model = get_model()
```

```python
callback = keras.callbacks.EarlyStopping(monitor='loss', min_delta=0, patience=1)
model_history = model.fit(X_train, y_train, epochs=200, batch_size=64, validation_data=(X_test, y_test), callbacks=[callback])
```

```python
def validate_model(model: keras.models.Sequential, X_test: np.ndarray, y_test: np.ndarray):
    y_pred = model.predict(X_test)
    #Converting predictions to label
    pred = [np.argmax(y_pred[i]) for i in range(len(y_pred))]
    test = [np.argmax(y_test[i]) for i in range(len(y_test))]
    
    a = accuracy_score(pred, test)
    print('Accuracy is:', a * 100)

    # plot accuracy
    plt.plot(model_history.history['accuracy'])
    plt.plot(model_history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # plot loss
    plt.plot(model_history.history['loss']) 
    plt.plot(model_history.history['val_loss']) 
    plt.title('Model loss') 
    plt.ylabel('Loss') 
    plt.xlabel('Epoch') 
    plt.legend(['Train', 'Test'], loc='upper left') 
    plt.show()
    
    return
```

```python
validate_model(model, X_test, y_test)
```
